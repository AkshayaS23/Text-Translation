```python
# Install required libraries
!pip install gTTS googletrans==4.0.0-rc1

from gtts import gTTS
from googletrans import Translator
from IPython.display import Audio, display
import os

# Initialize the Translator
translator = Translator()

def translate_text(text, dest_lang):
    """Translates the given text to the desired language."""
    translated = translator.translate(text, dest=dest_lang)
    return translated.text

def text_to_speech(text, lang='en', filename='output.mp3'):
    """Converts the given text to speech and saves the audio file."""
    tts = gTTS(text=text, lang=lang)
    tts.save(filename)
    audio = Audio(filename, autoplay=True)
    print(f"Download your audio file: {filename}")
    return audio

def is_valid_language(lang):
    """Check if the language code is valid for gTTS."""
    valid_languages = [
        'en', 'es', 'fr', 'de', 'it', 'pt',
        'ja', 'ko', 'hi', 'zh-cn', 'zh-tw', 'ta'
    ]
    return lang in valid_languages

# Example Usage
text_input = input("Enter the text you want to convert to speech: ")

# Select the language for translation
print("Supported language codes: 'en' for English, 'es' for Spanish, 'fr' for French, 'de' for German, 'hi' for Hindi, 'ta' for Tamil, etc.")
dest_lang = input("Enter the language code to translate the text (e.g., 'es' for Spanish, 'fr' for French, 'ta' for Tamil): ")

# Translate the text
translated_text = translate_text(text_input, dest_lang)
print(f"Translated Text ({dest_lang}): {translated_text}")

# Get the language for text-to-speech
tts_lang = input("Enter the language code for text-to-speech (default is the translation language): ") or dest_lang

# Validate the language
if not is_valid_language(tts_lang):
    print(f"Invalid language code '{tts_lang}'. Defaulting to English.")
    tts_lang = 'en'  # Fallback to English

# Get the filename (optional)
filename_input = input("Enter the filename (default is 'output.mp3'): ") or 'output.mp3'

# Convert the translated text to speech and display the audio
audio_file = text_to_speech(translated_text, lang=tts_lang, filename=filename_input)
display(audio_file)
```
